# Udemy courses 

# importing libraries

```python
import pandas as pd
import numpy as np

import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
%matplotlib inline
```

# importing dataset

```python
df= pd.read_csv('udemy_courses-raw.csv')
```

```python
df.head(2)
```

```python
df.info()
```

# handling columns

```python
# changing the course id column type to obj instead of integer because it is not considered as quantitative data type

df['course_id'] = df['course_id'].astype(str)
```

# checking for missing values

```python
df.isnull().sum()
```

# checking and handling duplicates

```python
df[df.duplicated(keep=False)].sort_values(by='course_id')
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

# handling outliers

```python
# exploring dataset for outliers

df.boxplot(figsize = (15,10))
plt.show()
```

```python
# dropping the extreme values from number of subscribers column 
# keeping the outliers from other columns because they are not too far off

df= df[df['num_subscribers'] < 100000 ]
```

```python
df.info()
```

# Exploratory data analysis

# 1. data distribution

Q1: what is the distribution of courses by subject?

the below graph shows the distribution of courses by subject where :
- Business finance comes first
- graphic design is next with only 1% less
- musical instruments is next by 19%
- web development is the least with only 13 %

```python
data = df['subject'].value_counts().tolist()
labels = df['subject'].unique()

#define Seaborn color palette to use
colors = sns.color_palette('pastel')[0:5]

#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%', radius= 3)
plt.show()
```

Q2: What is the distribution of courses by level?

the below graph shows that:
- the majority of courses are open for all levels
- 35% of courses are for intermmediate level
- 11% are for beginners only
- only 2% are for experts

```python
data1 = df['level'].value_counts().tolist()
labels1 = df['level'].unique()

#define Seaborn color palette to use
colors = sns.color_palette('pastel')[0:5]

#create pie chart
plt.pie(data1, labels = labels1, colors = colors, autopct='%.0f%%', radius= 3)
plt.show()
```

Q3: Are most of the courses paid or not ?

the graph below shows that most courses on the platform are paid

```python
data2 =df['is_paid']
plot = sns.countplot(x = data2,  data = df)
```

Q4 : How are the courses distributed by price categories?

the price categories are distributed excluding the free courses as follow :
    
- less than 50
- between 50 and 100 
- between 100 and 150
- between 150 and 200

the graph below shows that:

- most courses are most courses are less than 50 dollars
- only around 500 courses seem to be over a 100 dollars

```python
df['price'].describe()
```

```python
categories = ['less than 50', 'between 50 and 100', 'between 100 and 150', 'between 150 and 200']
amounts = [1,50,100,150,200]

# creating a new column for the new categories

df['category'] =  pd.cut(df['price'], bins= amounts, labels= categories, right=False)


# visualizing results

data1 =  df['category'].value_counts().tolist()
labels1 = categories

colors = sns.color_palette('pastel')[0:5]

plt.figure(figsize=(10, 10))
plt.bar(labels1, data1)
plt.xlabel('Category')
plt.ylabel('Price')
plt.title(' Distribution by price')
plt.grid(True, linewidth= 1, linestyle="--")
plt.show()
```

```python
df.drop('category', axis = 1, inplace = True )
```

Q5: what is the course  distribution by popularity?

the courses are divided by popularity as follow :
- unpopular : up to 1000 subscribers
- some what known : between 1000 and 3000 subscribers
- known: between 3000 and 5000 subscribers
- popular: between 5000 and 8000
- the most popular : between 8000 and 10000

the below graph shows most courses are unpopular with less than 1000 subscribers

```python
df['num_subscribers'].describe()
```

```python
popularity = ['unpopular', 'some what known', 'known', 'popular', 'the most popular']
subs = [0,1000,3000,5000,8000,10000]

df['popularity'] =  pd.cut(df['num_subscribers'], bins= subs, labels= popularity, right=False)


data2 =  df['popularity'].value_counts().tolist()
labels2 = popularity

colors = sns.color_palette('pastel')[0:5]

plt.figure(figsize=(10, 10))
plt.bar(labels2, data2)
plt.xlabel('popularity')
plt.ylabel('subscribers')
plt.title(' Distribution by popularity')
plt.grid(True, linewidth= 1, linestyle="--")
plt.show()
```
